```python
import jax.nn as jnn
import jax.numpy as jnp
from functools import partial

import matplotlib.pyplot as plt

gelu = partial(jnn.gelu, approximate=True)

def gelu_quad(x):
    ret = 0.125 * jnp.square(x) + 0.25 * x + 0.5
    return ret

xs = jnp.linspace(-5, 5, 50)

y_gelu = gelu(xs)
y_gelu_quad = gelu_quad(xs)

plt.plot(xs, y_gelu, label="gelu")
plt.plot(xs, y_gelu_quad, label="gelu_quad")

plt.legend()
plt.show()
```
